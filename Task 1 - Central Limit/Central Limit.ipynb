```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
population = np.random.randint(0, 30,10000)
```

```python
plt.figure(figsize=(10,10))
plt.hist(population,bins=50,density = True)
plt.show()
```

```python
def central_limit(samples_number,sample_size,population):
    """
    * description
    get the central limit distribution of the population
    
    * the function takes the follwoing parameters :
        - samples_number which is the number of sample you want to get from the population
        - sample size which is the size of samples
        - population that you wznt to get the central limit of it
    """
    
    # define a list to save samples means in it
    central_limit = []
    for i in range(samples_number):
        # get a random number between 0,1000
        random_number = np.random.randint(0,1000)
        # get a sample between random_number - sample_size and random_number 
        # then gets its mean
        mean_of_central = np.mean(population[random_number-sample_size:random_number])
        # add sample mean to the list
        central_limit.append(mean_of_central)
    # plot the distributaion of samples mean
    plt.hist(central_limit)
    plt.show()
    
```

```python
central_limit(10,30,population)
```

```python
central_limit(50,30,population)
```

```python
central_limit(100,30,population)
```

```python
central_limit(200,30,population)
```
